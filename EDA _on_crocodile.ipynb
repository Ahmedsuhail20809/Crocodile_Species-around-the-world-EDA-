Observation:
>Note: Before starting EDA, Everybody should know the meanings of each feature in the given dataset.
***Features purpose:***

- **Observation** **ID** → Unique number/code given to each crocodile record.

- **Common** **Name** → Everyday name (e.g., Nile crocodile).

- **Scientific** **Name** → Formal Latin species name (e.g., Crocodylus niloticus).

- ****Family**** → The biological family grouping (e.g., Crocodylidae).

- **Genus**→ The narrower biological category before species (e.g., Crocodylus).

- **Observed Length (m)** → Measured body length in meters.

- **Observed Weight (kg)** → Measured body weight in kilograms.

- **Age** **Class** → Life stage (juvenile, sub-adult, adult).

- **Sex** → Male or female.

- **Date** of **Observation** → When the sighting/measurement was recorded.

- **Country**/**Region** → Location of observation.

- **Habitat** **Type** → Environment where it was found (river, lake, swamp).

- **Conservation** **Status** → IUCN Red List or other status (e.g., Vulnerable).

- **Observer** **Name** → Person who recorded the observation.

- ****Notes**** → Extra remarks or special conditions noticed.

# 3.2 Find out outliers:

```python

Q1 = df['Observed Weight (kg)'].quantile(0.25)
Q3 = df['Observed Weight (kg)'].quantile(0.75)
IQR = Q3 -Q1
print('This is the value of IQR=',IQR)

# let's find upper limit and lower limit

lower_limit = Q1 -1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print('The lower limit is',lower_limit)
print('The upper limit is',upper_limit)

# let's detect outliers
outliers = df[(df['Observed Weight (kg)'] < lower_limit) | (df['Observed Weight (kg)'] > upper_limit)]
print('The outliers are',outliers)
```

```python
# let's remove the outliers
df = df[(df['Observed Weight (kg)'] > lower_limit) & (df['Observed Weight (kg)'] < upper_limit)]
```

```python
sns.histplot(data =df,x=df['Observed Weight (kg)'],kde=True)
```

```python
# let's plot interactive plot by using plotly
import plotly.express as px

# Convert 'Date of Observation' to datetime
df['Date of Observation'] = pd.to_datetime(df['Date of Observation'])

# Extract year
df['Year'] = df['Date of Observation'].dt.year

# Group by year and count records
yearly_counts = df.groupby('Year').size().reset_index(name='Records')

# Draw interactive line plot
fig = px.line(yearly_counts, x='Year', y='Records', title='Yearly Crocodile Data Collection')
fig.show()
```

```python
fig = px.sunburst(
    df,
    path=['Common Name', 'Sex', 'Country/Region', 'Observer Name'],
    title='Crocodile Species Sunburst: Common Name, Sex, Country/Region, Observer Name',
    maxdepth=-1
)
fig.show()
```
